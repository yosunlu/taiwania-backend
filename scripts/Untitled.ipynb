```python
import pandas as pd
import requests
```

```python
df = pd.read_csv('/Users/yushanlu/Desktop/yushan-says/backend/scripts/data.csv', sep='|')
df
```

```python
# Add a column, separating slang/everyday language and other tags
df.insert(4, "usage", "")
df
```

```python
# Add values for column usage
for i in range(len(df)):
    if df.at[i, 'id'] <= 380:
        df.at[i, 'usage'] = "Proverb"    
    else:
        df.at[i, 'id'] = i + 1
        df.at[i, 'usage'] = "Everday Language"  
    df.at[i, 'tags'] = ""
df
    
```

```python
# Categorize the phrases
for i in range(len(df)):
    phrase = df.at[i, 'usage'] + ": " + df.at[i, 'phrase']
    # print(phrase)
    
    response = requests.post(
        "http://localhost:8000/categorize/invoke",
        json = {
            'input': {'phrase' : phrase}}
    )
    tags = response.json()['output']['content']
    # print(tags)
    df.at[i, 'tags'] = tags
    
```

```python
df
```

```python
# Add url for Everyday Lanuguar audiourl
for i in range(380, 399):
    df.at[i, 'audiourl'] = "https://itaigi.tw/k/" + df.at[i, 'phrase'][:-1]
df
```

```python
df.insert(3, "mandarin", "")
df
```

```python
# Add mandarin translation 
for i in range(len(df)):
    phrase = df.at[i, 'phrase']
    definition = df.at[i, 'definition']
    response = requests.post(
        "http://localhost:8000/translate-to-mandarin/invoke",
        json = {
            'input': {'phrase' : phrase, 'definition': definition}}
    )
    df.at[i, 'mandarin'] = response.json()['output']['content']

df
```
